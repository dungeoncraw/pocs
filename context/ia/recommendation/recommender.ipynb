This is a simple recommender based on https://www.kaggle.com/datasets/rounakbanik/the-movies-dataset
Download the dataset and save under "source" folder

```python
import pandas as pd
import numpy as np
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import linear_kernel

# Fix dtype warnings by specifying column types
movies = pd.read_csv(
    './source/movies_metadata.csv',
    dtype={
        'id': str,          # Handle non-integer IDs
        'popularity': str   # Clean later
    },
    low_memory=False
)

# Clean 'id' column (remove non-numeric characters)
movies['id'] = movies['id'].str.replace(r'\D', '', regex=True)
movies = movies[movies['id'].str.len() > 0]  # Drop empty IDs
movies['id'] = movies['id'].astype(int)

# Keep essential columns
movies = movies[['id', 'title', 'genres', 'overview']]
movies.dropna(subset=['overview'], inplace=True)  # Remove movies without overviews
```

```python
# Combine genres and overview into a single text feature
def preprocess_text(row):
    # Extract genres (e.g., "Animation, Comedy, Family")
    genres = eval(row['genres'])
    genres_str = ', '.join([g['name'] for g in genres]) if genres else ''
    
    # Repeat genres 3x to increase their weight in TF-IDF
    return f"{genres_str} {genres_str} {genres_str} {row['overview']}"

movies['text'] = movies.apply(preprocess_text, axis=1)
```

```python
# Create TF-IDF Embeddings
tfidf = TfidfVectorizer(
    stop_words='english',
    ngram_range=(1, 2),      # Include bigrams (e.g., "Science Fiction")
    max_features=5000        # Focus on most impactful terms
)
tfidf_matrix = tfidf.fit_transform(movies['text'])
tfidf_matrix = tfidf.fit_transform(movies['text'])


# Compute Cosine Similarities
cosine_sim = linear_kernel(tfidf_matrix, tfidf_matrix)

# Build Recommendation Function
def get_recommendations(title, num_recommendations=5):
    # Get movie index
    idx = movies[movies['title'] == title].index[0]
    
    # Get similarity scores
    sim_scores = list(enumerate(cosine_sim[idx]))
    
    # Sort by score
    sim_scores = sorted(sim_scores, key=lambda x: x[1], reverse=True)
    
    # Get top N recommendations (skip self)
    sim_scores = sim_scores[1:num_recommendations+1]
    movie_indices = [i[0] for i in sim_scores]
    
    return movies['title'].iloc[movie_indices]
```

```python
def get_filtered_recommendations(title, primary_genre="Animation", num_recommendations=5):
    # Get raw recommendations
    idx = movies[movies['title'] == title].index[0]
    sim_scores = list(enumerate(cosine_sim[idx]))
    sim_scores = sorted(sim_scores, key=lambda x: x[1], reverse=True)
    
    # Filter movies that include the primary genre
    filtered_indices = []
    for i, score in sim_scores:
        genres = eval(movies.iloc[i]['genres'])
        genre_names = [g['name'] for g in genres]
        if primary_genre in genre_names and i != idx:  # Skip the query movie
            filtered_indices.append(i)
        if len(filtered_indices) == num_recommendations:
            break
    
    return movies.iloc[filtered_indices]['title']
```

```python
# Test Recommendations
get_filtered_recommendations("Toy Story", primary_genre="Animation")

import matplotlib.pyplot as plt
import seaborn as sns

# Plot similarity scores for a specific movie
example_idx = movies[movies['title'] == "Toy Story"].index[0]
similarities = cosine_sim[example_idx]

plt.figure(figsize=(10, 5))
sns.barplot(x=similarities, y=movies['title'], orient='h', order=movies.iloc[np.argsort(similarities)[-10:]]['title'])
plt.title("Top 10 Similar Movies to 'Toy Story'")
plt.show()
```
